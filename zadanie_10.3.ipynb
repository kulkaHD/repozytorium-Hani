```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats

# przygotowanie danych
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

# Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej. Opisz różnice pomiędzy statystkami i wynik swojej pracy wyślij Mentorowi jako Jupyter Noteboook zapisany w Github.

```python
zbior_100_elementow_rozklad_normalny = np.sort(stats.norm.rvs(3, 1, 100)).round(2)
zbior_10000_elementow_rozklad_normalny = np.sort(stats.norm.rvs(3, 1, 10000)).round(2)
```

```python
x100 = np.linspace(-10, 10, 100)
rozklad_prawdopodobienstwa_dla_100_elementow_rn = stats.norm.pdf(x100, 0, 1)
x10000 = np.linspace(-10, 10, 10000)
rozklad_prawdopodobienstwa_dla_10000_elementow_rn = stats.norm.pdf(x10000, 0, 1)

plt.figure(figsize=(3, 2))
plt.plot(zbior_100_elementow_rozklad_normalny, rozklad_prawdopodobienstwa_dla_100_elementow_rn, label='zbiór 100 elementów o rozkładzie normalnym (odchylenie standardowe = 1, średnia = 3)')
plt.legend(loc=(1.05,0.9))
plt.show()

plt.figure(figsize=(3, 2))
plt.plot(zbior_10000_elementow_rozklad_normalny, rozklad_prawdopodobienstwa_dla_10000_elementow_rn, label='zbiór 10 000 elementów o rozkładzie normalnym (odchylenie standardowe = 1, średnia = 3)')
plt.legend(loc=(1.05,0.9))
plt.show()
```

Z powodu większej ilości punktów wykres zbioru 10 000-elementowego jest bardziej zbliżony do ideału odwzorowania rozkładu normalnego.

# średnia

```python
srednia_rn100 = np.round(np.mean(zbior_100_elementow_rozklad_normalny),2)
srednia_rn10000 = np.round(np.mean(zbior_10000_elementow_rozklad_normalny),2)
print(f'średnia dla x=100 wynosi {srednia_rn100}')
print(f'średnia dla x=10000 wynosi {srednia_rn10000}')
```

Im więcej elementów posiada zbiór, tym bardziej średnia zbliża się do tej zadanej wynoszącej 3.

# mediana

```python
mediana_rn100 = np.round(np.median(zbior_100_elementow_rozklad_normalny),2)
mediana_rn10000 = np.round(np.median(zbior_10000_elementow_rozklad_normalny),2)
print(f'mediana dla x=100 wynosi {mediana_rn100}')
print(f'mediana dla x=10000 wynosi {mediana_rn10000}')
```

Mediana w rozkładzie normalnym wynosi tyle,co średnia. Na powyższym przykładzie widać,że różnice między średnią a medianą są tym mniejsze, im bardziej liczebny jest zbiór.

# moda

```python
moda_rn100 = stats.mode(zbior_100_elementow_rozklad_normalny)
moda_rn10000 = stats.mode(zbior_10000_elementow_rozklad_normalny)
print(f'moda dla x=100 wynosi {moda_rn100}')
print(f'moda dla x=10000 wynosi {moda_rn10000}')
```

Moda w rozkładzie normalnym wynosi tyle,co średnia. Na powyższym przykładzie widać,że różnice między średnią a modą są tym mniejsze, im bardziej liczebny jest zbiór.

# kwartyle

```python
q0_rn100 = np.quantile(zbior_100_elementow_rozklad_normalny, 0.0)
q1_rn100 = np.round(np.quantile(zbior_100_elementow_rozklad_normalny, 0.25),2)
q3_rn100 = np.round(np.quantile(zbior_100_elementow_rozklad_normalny, 0.75),2)
q4_rn100 = np.quantile(zbior_100_elementow_rozklad_normalny, 1.0)

q0_rn10000 = np.quantile(zbior_10000_elementow_rozklad_normalny, 0.0)
q1_rn10000 = np.round(np.quantile(zbior_10000_elementow_rozklad_normalny, 0.25),2)
q3_rn10000 = np.round(np.quantile(zbior_10000_elementow_rozklad_normalny, 0.75),2)
q4_rn10000 = np.quantile(zbior_10000_elementow_rozklad_normalny, 1.0)
print(f'''
Q0_rn100: {q0_rn100}
Q1_rn100: {q1_rn100}
Q2_rn100: {mediana_rn100}
Q3_rn100: {q3_rn100}
Q4_rn100: {q4_rn100}

Q0_rn10000: {q0_rn10000}
Q1_rn10000: {q1_rn10000}
Q2_rn10000: {mediana_rn10000}
Q3_rn10000: {q3_rn10000}
Q4_rn10000: {q4_rn10000}
''')
```

Kwartyle pierwsze i trzecie są do siebie mocno zbliżone w obu zbiorach.Kwartyle zerowe i czwarte ze względu na większy zakres zbioru 10 000-elementowego różnią się.

# zakres

```python
range_rn100 = max(zbior_100_elementow_rozklad_normalny) - min(zbior_100_elementow_rozklad_normalny)
range_rn10000 = max(zbior_10000_elementow_rozklad_normalny) - min(zbior_10000_elementow_rozklad_normalny)
print(f'''
zakres zbioru 100-elementowego wynosi {range_rn100}
zakres zbioru 10 000-elementowego wynosi {range_rn10000}
''')
```

Ze względu na większą liczebność zbioru elementy graniczne zbioru rn_10000 są od siebie bardziej oddalone i nie zaburza to zadanych wartości jak średnia=3 i odchylenie standardowe=1. 

# rozstęp międzykwartylowy - IQR

```python
iqr_rn100 = q3_rn100 - q1_rn100
iqr_rn10000 = q3_rn10000 - q1_rn10000
print(f'''
rozstęp międzykwartylowy dla zbioru 100-elementowego wynosi {iqr_rn100}
rozstęp międzykwartylowy dla zbioru 10 000-elementowego wynosi {iqr_rn10000}
''')
```

Im większa liczebność zbioru, tym wartość IQR jest bliższa wartości 1.34.

Nie umiem powiedzieć czym jest ta liczba, ale jest szczególna - badanie internetu wskazuje, że w rozkładzie normalnym z jakiegoś powodu wartość 1.34 jest istotna. Jednak za mały ze mne żuczek, żeby powiedzieć dlaczego.

# wykres pudełkowy

```python
plt.boxplot(zbior_100_elementow_rozklad_normalny, vert=False)
plt.show()
plt.boxplot(zbior_10000_elementow_rozklad_normalny, vert=False)
plt.show()
```

Im większa liczebność zbioru tym więcej wartości odstających, które nie zaburzają założeń zbioru (średnia = 3, odchylenie standardowe = 1)

# wariancja

```python
variance_rn100 = np.var(zbior_100_elementow_rozklad_normalny, ddof=1)
variance_rn10000 = np.var(zbior_10000_elementow_rozklad_normalny, ddof=1)
print(f'''
wariancja dla zbioru 100-elementowego wynosi {variance_rn100}
wariancja dla zbioru 10 000-elementowego wynosi {variance_rn10000}
''')
```

Im wieksza liczebność zbioru, tym wariancja bliższa jest wartości 1.

# odchylenie standardowe

```python
s_rn100 = variance_rn100**(1/2)
s_rn10000 = variance_rn10000**(1/2)
print(f'''
odchylenie standardowe dla zbioru 100-elementowego wynosi {s_rn100}
odchylenie standardowe dla zbioru 10 000-elementowego wynosi {s_rn10000}
''')
```

```python
standard_deviation_rn100 = np.std(zbior_100_elementow_rozklad_normalny, ddof=1)
standard_deviation_rn10000 = np.std(zbior_10000_elementow_rozklad_normalny, ddof=1)
print(f'''
odchylenie standardowe dla zbioru 100-elementowego wynosi {standard_deviation_rn100}
odchylenie standardowe dla zbioru 10 000-elementowego wynosi {standard_deviation_rn10000}
''')
```

Im wieksza liczebność zbioru, tym odchylenie standardowe bliższe jest zadanej wartości 1.
